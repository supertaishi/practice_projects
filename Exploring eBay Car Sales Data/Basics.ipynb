## Step2

Rename the column names from camelcase to snakecase.

```python
corrections = {
    'dateCrawled': 'date_crawled',
    'offerType': 'offer_type',
    'vehicleType': 'vehicle_type',
    'yearOfRegistration': 'registration_year',
    'powerPS': 'power_ps',
    'monthOfRegistration': 'registration_month',
    'fuelType': 'fuel_type',
    'notRepairedDamage': 'unrepaired_damage',
    'dateCreated': 'ad_created',
    'nrOfPictures': 'nr_of_pictures',
    'postalCode': 'postal_code',
    'lastSeen': 'last_seen'
}

autos.rename(columns = corrections, inplace=True)
autos.head()
```

## Step3

### Investigation
- 'seller' and 'offer_type' have only 2 unique values and 1 value appears only once. 'abtest', 'gearbox', and 'unrepaired_damage' have also only 2 unique values.
- 'price' and 'odometer' columns are numeric values stored as text.

Improve the 'price' and 'odometer' columns

```python
autos['price'] = autos['price'].str.replace('$', '').str.replace(',', '').astype(int)

# autos['price'].value_counts()
# autos.describe(include='all')
# autos.info()
```

```python
autos['odometer'] = autos['odometer'].str.replace('km', '').str.replace(',', '').astype(int)

# autos['odometer'].value_counts()
# autos.info()
```

```python
autos.rename(columns={'odometer': 'odometer_km'}, inplace=True)

# autos.info()
```

## Step7

```python
autos['brand'].unique()
len(autos['brand'].unique())
```

```python
brand_bool = autos['brand'].value_counts(normalize=True) > 0.01
major_brands = autos['brand'].value_counts(normalize=True).index[brand_bool]
```

Set the 'major_brands' as the brands whose shares are more than 1% of total.

The brands are ['volkswagen', 'bmw', 'opel', 'mercedes_benz', 'audi', 'ford', 'renault',
       'peugeot', 'fiat', 'seat', 'skoda', 'mazda', 'nissan', 'citroen',
       'smart', 'toyota', 'sonstige_autos']

```python
mean_prices = {}
for b in major_brands:
    one_mean = autos.loc[autos['brand'] == b, 'price'].mean()
    mean_prices[b] = np.round(one_mean, 1)
mean_prices
```

```python
mp_series = pd.Series(mean_prices)
print(mp_series)
df = pd.DataFrame(mp_series, columns=['mean_price'])
print(df)
```

```python
mean_mile = {}
mean_price = {}
for b in major_brands:
    one_mean_price = autos.loc[autos['brand'] == b, 'price'].mean()
    one_mean_mile = autos.loc[autos['brand'] == b, 'odometer_km'].mean()
    mean_price[b] = np.round(one_mean_price, 1)
    mean_mile[b] = np.round(one_mean_mile, 1)

print(mean_mile)
print()
print(mean_price)
```

```python
mm = pd.Series(mean_mile).sort_values(ascending=False)
print(mm)
print('\n')
mp = pd.Series(mean_price).sort_values(ascending=False)
print(mp)
```

```python
df = pd.DataFrame(mm, columns=['mean_mile'])
df['mean_price'] = mp
df
```
